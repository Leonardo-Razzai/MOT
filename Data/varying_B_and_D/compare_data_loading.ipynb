```python
B_vals = range(5, 12, 1)
color_palette = plt.cm.jet(np.linspace(0.0, 0.5, len(B_vals)))
DATE = './2025-01-16/'
```

```python
_, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.set_title('Fluorescence maxima at different B values', fontdict=title_font)
ax.set_xlabel('Loading BN Cooler Frequency [MHz]', fontdict=base_font)
ax.set_ylabel('Photoemission [V]', fontdict=base_font)
for i, B in enumerate(B_vals):
  df = pd.read_csv(DATE + f'Photoemission_Data_B={B}.csv')
  x_data = df['BN_cooler [MHz]']
  
  y_data = df['PhotoEm [V]']
  y_err = df['Error [V]']
  
  ax.errorbar(x_data, y_data, yerr=y_err, fmt='--o', label=f'B = {B} A', color=color_palette[i])

plt.grid()
plt.legend()
plt.show()
```

```python
_, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.set_title('Number of atoms at different B values', fontdict=title_font)
ax.set_xlabel('Loading BN Cooler Frequency [MHz]', fontdict=base_font)
ax.set_ylabel(r'Number of atoms [$10^7$]', fontdict=base_font)
for i, B in enumerate(B_vals):
  df = pd.read_csv(DATE + f'Photoemission_Data_B={B}.csv')
  x_data = df['BN_cooler [MHz]']
  
  y_data = df['PhotoEm [V]']
  y_err = df['Error [V]']
  
  num_atoms, err_num = NumOfAtoms(y_data, y_err, Delta=0, Gamma=13) # Gamma is irrelevant at Delta = 0
  
  ax.errorbar(x_data, num_atoms/1e7, yerr=err_num/1e7, fmt='--o', label=f'B = {B} A', color=color_palette[i])

plt.grid()
plt.legend()
plt.show()
```

```python
_, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.set_title('Loading times at different B values', fontdict=title_font)
ax.set_xlabel('Loading BN Cooler Frequency [MHz]', fontdict=base_font)
ax.set_ylabel(r'$\tau$ [s]', fontdict=base_font)
for i, B in enumerate(B_vals):
  df = pd.read_csv(DATE + f'Load_params_B={B}.csv')
  x_data = df['BN_cooler [MHz]']
  
  y_data = df['tau [s]']
  y_err = df['dtau [s]']
    
  ax.errorbar(x_data, y_data, yerr=y_err, fmt='--o', label=f'B = {B} A', color=color_palette[i])

plt.grid()
plt.legend()
plt.show()
```
